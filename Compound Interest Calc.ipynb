```python
#Compund Interest Formula

# P = Initial Principal Balance
P = input('Initial Principal Balance ($): ')
P = int(P)

# r = Interest Rate
r = input('Interest Rate to the 1/100th: ')
r = float(r)

# n = Number of times Interest applied per time period
n = input('Number of times Interest applied per time period: ')
n = int(n)

# t = Number of time periods elasped
t = input('Number of time periods elapsed: ')
t = int(t)

# A = Final Amount
A = P*(1+(r/n))**(n*t)
A = round(A,2)
print('Final Amount: ' + str(A))
```
